# Descriptive Analysis of the Compiled HRS data 

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

PATH_TO_DATA = Path("./HRS_3Darray.npy")
```

```python
data = np.load(PATH_TO_DATA)
varnames = ['age', 
            'gender', 
            'education', 
            'month', 
            'day', 
            'year', 
            'day_week', 
            'paper', 
            'cactus', 
            'president', 
            'vice_president',
            'serie_7',
            'backwards',
            'recall_immediate',
            'recall_delayed']
```

```python
def plotHRS(data, obs=range(100), var=range(3,4)):
    fig, ax = plt.subplots(nrows=1, ncols=len(var))
    for v in range(len(var)):
        ax[v].imshow(data[obs, var[v], :], cmap="viridis")
        ax[v].set_title(f'Plot of {varnames[var[v]]}')
    plt.subplots_adjust(hspace=.5)
    plt.show()
```

```python
print('varname\t\t\t', 'min\t', 'max\t', 'mean')
print((50*'-')[:45])
for i in range(data.shape[1]):
    print((varnames[i] + 15*' ')[:16],  '\t', np.nanmin(data[:,i]), '\t', np.nanmax(data[:,i],), '\t', np.round(np.nanmean(data[:,i]), 2))
```

```python
plotHRS(data, obs=range(100), var=range(4))
```

```python
plotHRS(data, obs=range(100), var=range(4,8))
```
